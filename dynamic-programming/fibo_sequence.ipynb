```python
import time
```

```python
# o(2n) solution
# the time increase exponentially
def fibo(n):
  result = 0;
  if(n==1 or n==2):
    result = 1
  for i in range(1,n):
    result = fibo(n-1) + fibo(n-2)
  return result

```

```python
start = time.perf_counter()
print(fibo(10))
end = time.perf_counter()
print(f"Execution time: {end - start:.6f} seconds")
```

```python
# O(n) solution
# storing the calculated values to save computation time (memoization)
# slow for big numbers of n
def fibo_memo(n, memo = None):
  if(memo is None):
    memo = {}
  if n==1 or n==2 :
    result = 1
  else:
    result = fibo_memo(n-1) + fibo_memo(n-2)
  memo[n] = result
  return result
```

```python
start = time.perf_counter()
print(fibo_memo(14))
end = time.perf_counter()
print(f"Execution time: {end - start:.6f} seconds")
```

```python
# O(n) - loop through the array only once
# bottom up solution
# we build an array to store the previous numbers in fibo sequence and speed up the calculation
def fibo_bottom_up(n):
  if(n==1 or n==2):
    result = 1
  bottom_up = [None] * (n+1)
  bottom_up[1] = 1
  bottom_up[2] = 1
  for i in range(3,n+1):
    bottom_up[i] = bottom_up[i-1] + bottom_up[i-2]
  return bottom_up[n]
```

```python
start = time.perf_counter()
print(fibo_bottom_up(14))
end = time.perf_counter()
print(f"Execution time: {end - start:.6f} seconds")
```
